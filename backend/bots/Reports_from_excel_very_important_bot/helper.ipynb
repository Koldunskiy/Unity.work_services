```python
def get_report(path:str):
    df = pd.read_excel(path)
    date, balanc = df.columns

    df[date] = pd.to_datetime(df["Дата"])
    df.set_index(date, inplace=True)

    # Вычислим доходность на основе изменения баланса
    df['Доходность'] = df[balanc].pct_change()

    # Удаляем строки с NaN в доходности
    returns = df['Доходность'].dropna()

    # (benchmark can be a pandas Series or ticker)
    qs.reports.html(returns, output="quantstats_report.html")

get_report()
```
