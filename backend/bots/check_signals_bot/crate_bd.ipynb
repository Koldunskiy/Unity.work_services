```python
import sqlite3

def create_database():
    # Подключение к базе данных (или её создание, если не существует)
    conn = sqlite3.connect('trading_analysis.db')
    cursor = conn.cursor()

    # Создание таблицы
    cursor.execute('''
    CREATE TABLE IF NOT EXISTS Trade_Signals (
        id INTEGER PRIMARY KEY AUTOINCREMENT,
        instrumentId varchar(30) NOT NULL,
        analyseType varchar(30) NOT NULL,
        header TEXT,
        body TEXT NOT NULL,
        createTime varchar(26),
        foreCast varchar(10),
        entry REAL,
        tp1 REAL,
        tp2 REAL,
        sl REAL,
        is_check INTEGER DEFAULT 0
    );
    ''')

    # Сохранение изменений и закрытие соединения
    conn.commit()
    conn.close()

# Вызов функции для создания базы данных и таблицы
create_database()
```
